```python
#Data Collection
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
import pandas as pd

# load the dataset

data = gutenberg.raw('shakespeare-hamlet.txt')

#save data to file
with open('hamlet.txt', 'w') as file:
    file.write(data)
```

```python
# Data Preprocessing

import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

#load the dataset
with open('hamlet.txt', 'r') as file:
    text=file.read().lower()

#Tokenize the text-creating indexes for words
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1
total_words
```

```python
tokenizer.word_index
```

```python
# Create input sequences
input_sequences = []
for line in text.split('\n'):
    token_list=tokenizer.texts_to_sequences([line])[0]
    for i in range(0, len(token_list)):
        n_grams_sequence = token_list[:i+1]
        input_sequences.append(n_grams_sequence)
```

```python
input_sequences # sentences are converted into input sequence
```

```python
# Applying Pad Sequence
max_sequence_len = max([len(x) for x in input_sequences]) #consider max sentence length 
max_sequence_len
```

```python
input_sequences=np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))
input_sequences # All sentences are same size
```

```python
# create predictors and label
import tensorflow as tf
x,y = input_sequences[:,:-1], input_sequences[:,-1]
```

```python
x
```

```python
y
```

```python
y=tf.keras.utils.to_categorical(y,num_classes=total_words) # converting output features to catergories
y
```

```python
#Splitting dataset into Train and Test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Train LSTM RNN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout

#Define the model
model = Sequential()
model.add(Embedding(total_words,100,input_length=max_sequence_len-1)) # 100 is number of dimensions for embeddings
model.add(LSTM(150,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words, activation="softmax"))

#compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# Define early stopping
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=350, restore_best_weights=True)
```

```python
# Train the model

history = model.fit(x_train, y_train, epochs=400, verbose=1, validation_data=(x_test, y_test), callbacks=[early_stopping])
```

```python
#Function to predict the next word
def predict_next_word(model, tokenizer, text, max_sequence_len):
    token_list = tokenizer.texts_to_sequences([text])[0]
    if len(token_list) >= max_sequence_len:
        token_list = token_list[-(max_sequence_len-1):]
    token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
    predicted=model.predict(token_list, verbose=0)
    predicted_word_index = np.argmax(predicted, axis=1)
    for word, index in tokenizer.word_index.items():
        if index == predicted_word_index:
            return word

    return None
```

```python
input_text = "But to recouer of vs by strong hand And termes"
print(f"input text:{input_text}")
max_sequence_len=model.input_shape[1]+1
next_word=predict_next_word(model, tokenizer, input_text, max_sequence_len)
print(f"Next word prediction:{next_word}")
```
